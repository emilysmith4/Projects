## Shot Distance

```
px.histogram(twos['shot_distance'],title='Two Point Shot Distance')
```

```
acc_df = pd.DataFrame(columns=['shot_distance','accuracy','volume'])
grouped = twos.groupby(by=['shot_distance','event_type']).count().reset_index()
i = 0
for dist in grouped['shot_distance'].unique():
    filt = grouped[grouped['shot_distance']==dist]
    filt = filt.set_index('event_type')
    made = filt.loc['made two']['id']
    miss = filt.loc['miss two']['id']
    acc = made/(made+miss)
    vol = made+miss
    acc_df.loc[i] = [dist,acc,vol]
    i+=1
```

```
px.scatter(acc_df,x='shot_distance',y='accuracy',size='volume',title='Two Point Accuracy by Distance')
```

Taking into consideration shooting fouls

```
foul = pbp[pbp["event_type"]!='timeout']
foul = foul[foul["event_type"]!='team rebound']
foul = foul[foul["event_type"]!='tech foul']
foul = foul[foul["event_type"]!='violation']
foul = foul[foul["event_type"]!='NONE']
foul = foul[foul["event_type"]!='replay']
foul = foul[~foul['visitor_description'].str.contains('Technical')]
foul = foul[~foul['home_description'].str.contains('Technical',na=False)]
foul = foul.reset_index()
```

```
two_a, three_a = 0, 0
fouls = foul[foul['event_type']=='shoot foul']
for i in fouls.index:
    inc = 1
    go = 1
    ft = 0
    back = 1
    #See what follows
    while go:
        next_event = foul.loc[i+inc]['event_type']
        if next_event in ['miss ft','make ft']:
            ft+=1
            inc+=1
        else:
            go=0
    if ft==2:
        two_a+=1
    elif ft==3:
        three_a+=1
```

```
dist = pbp['event_type'].value_counts().to_frame('counts')
dist.loc['miss two'] = dist.loc['miss two'] + two_a
dist.loc['miss three'] = dist.loc['miss three'] + three_a
dist.loc['two attempt'] = dist.loc['made two']+dist.loc['miss two']
dist.loc['three attempt'] = dist.loc['made three']+dist.loc['miss three']
dist.loc['ft attempt'] = dist.loc['make ft']+dist.loc['miss ft']
dist = dist.drop(index=['made two','made three','miss two','miss three','make ft','miss ft'])
dist = dist.reset_index().rename(columns={'index':'event'})
dist['proportion'] = dist['counts']/(dist['counts'].sum())
dist = dist.sort_values(by='counts',ascending=False)
px.bar(dist, x='event', y='proportion',title='Counting Shots with a Foul')
```

## Next Event

```
pbp = pbp.reset_index()
```

```
def next_event(event):
    lst = []
    all_events = ['end quar','replay','NONE']
    rebound = ['def rebound','off rebound', 'team rebound']
    shot = ['miss two', 'made two', 'miss three', 'made three']
    ft = ['make ft', 'miss ft']
    fouls = ['shoot foul', 'def foul','tech foul']
    if event in ['miss two', 'miss three']:
        check = all_events+shot+ft
    elif (event in fouls) or (event=='make ft'):
        check = all_events+rebound
    elif event == 'miss ft':
        check = all_events+shot+['team rebound']+fouls
    else:
        check = all_events+rebound+ft

    for i in pbp.index:
        event_type = pbp.loc[i]['event_type']
        if event_type == event:
            if event in ['off foul','flag foul']:
                next_event = pbp.loc[i+2]['event_type']
            else:
                next_event = pbp.loc[i+1]['event_type']
            if next_event not in check:
                lst.append(next_event)

    series = pd.Series(lst).value_counts().to_frame('counts')
    if event not in ['miss two', 'miss three','shoot foul','miss ft','flag foul']:
        series.loc['two attempt'] = series.loc['made two']+series.loc['miss two']
        series.loc['three attempt'] = series.loc['made three']+series.loc['miss three']
        series = series.drop(index=['made two','made three','miss two','miss three'])
    series = series.reset_index().rename(columns={'index':'event'})
    series['proportion'] = series['counts']/(series['counts'].sum())
    series = series.sort_values(by='counts',ascending=False)
    #px.bar(series, x='event', y='proportion')
    return series
```

```
event = 'made two'
props = next_event(event)
px.bar(props, x='event', y='proportion',title='Probability of events following '+event)
```

## Shooting Foul

```
foul = pbp[pbp["event_type"]!='timeout']
foul = foul[foul["event_type"]!='team rebound']
foul = foul[foul["event_type"]!='tech foul']
foul = foul[foul["event_type"]!='violation']
foul = foul[foul["event_type"]!='NONE']
foul = foul[foul["event_type"]!='replay']
foul = foul[~foul['visitor_description'].str.contains('Technical')]
foul = foul[~foul['home_description'].str.contains('Technical',na=False)]
foul = foul.reset_index()
```

```
two_a, three_a, two_m, three_m = 0, 0, 0, 0
else_lst = []
fouls = foul[foul['event_type']=='shoot foul']
for i in fouls.index:
    inc = 1
    go = 1
    ft = 0
    back = 1
    #See what follows
    while go:
        next_event = foul.loc[i+inc]['event_type']
        if next_event in ['miss ft','make ft']:
            ft+=1
            inc+=1
        else:
            go=0
    if ft==1:
        prev_event = foul.loc[i-1]['event_type']
        if 'two' in prev_event:
            two_m+=1
        if 'three' in prev_event:
            three_m+=1
    elif ft==2:
        two_a+=1
    elif ft==3:
        three_a+=1
    else:
        else_lst.append(i)
```

# Game Analysis

```
boxscores = pd.read_csv('boxscores.csv')
```

```
boxscores.columns
```

```
def boxscore_explore(stat, function, differential=0):
    if not differential:
        if function ==  px.histogram:
            return function(boxscores[stat],labels={'value':str(stat)})
        else:
            return function(boxscores[stat])
    else:
        lst = []
        for i in boxscores['GAME_ID'].unique():
            game = boxscores[boxscores['GAME_ID']==i].reset_index()[stat]
            team1 = game[0]
            team2 = game[1]
            lst.append(abs(team1-team2))
        if function == px.histogram:
            return function(lst,labels={'value':str(stat)+' differential'})
        else: 
            return function(lst)
```

```
boxscore_explore('PTS',px.histogram, 1)
```
